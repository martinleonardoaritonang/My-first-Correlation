```python
# Martin Leonardo Aritonang
# 10219092
# Korelasi antar kolom

#import library yang diperlukan

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(10,4)

data_movies = pd.read_csv('movies_dataset.csv')
data_movies.shape
```

```python
#Cek dataset apakah ada null atau tidak
data_movies.isnull().values.any()

for col in data_movies.columns :
    missing = np.mean(data_movies[col].isnull())
    print('{} - {}%'.format(col, round(missing,5)))
    
#apabila data null terlalu besar makan gunakan perintah dibawah

# data_movies_without_null = data_movies.dropna()
```

```python
#Eksplor data

data_movies.head()
```

```python
data_movies = data_movies.sort_values(by=['budget'], inplace=False, ascending=False)
```

```python
#melihat apakah ada data yang sama pada kolom company

data_movies['company'].drop_duplicates().sort_values(ascending=False)
```

```python
data_movies.groupby('company').mean()
```

```python
data_movies.head()
```

```python
# korelasi 1
# mencari hubungan budget dengan gross

plt.scatter(x=data_movies['budget'], y =data_movies['gross'])

plt.title('budget vs gross')

plt.xlabel('budget for film')

plt.ylabel('gross earnings')

plt.show()
```

```python
# budget vs gross using seaborn

sns.regplot(x='budget', y='gross',data=data_movies,scatter_kws={'color':'black'}, line_kws={'color':'blue'})
```

```python
# metode korelasi ada pearson, kendall, dan spearman
# pearson : mencari hubungan antara dua variabel, yaitu variabel bebas dan variabel terikat
# kendall : mencari hubungan antara dua variabel atau lebih dan datanya bersifat ordinal
# spearman : mencari hubungan antara dua variabel atau lebih dan datanya bersifat ordinal
# sumber data Kendall berasal dari subjek yang sama atau berpasangan
# sumber data spearman berasal dari subjek yang berbeda

pearson_correlation_movies = data_movies.corr(method='pearson') 
```

```python
# kita plot kendall correlation movies kedalam heatmap

sns.heatmap(pearson_correlation_movies, annot=True)

plt.title('Pearson correlation movies')

plt.xlabel('name of columns')

plt.ylabel('name of columns')

plt.show()
```
